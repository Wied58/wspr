```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
wspr = pd.read_csv('cleaned.csv')
wspr.shape
```

```python
wspr.drop_duplicates(subset ="time", keep = False, inplace = True)
wspr.shape
```

```python
wspr
```

```python

#df['col'] = df['col'].str.slice(0, 9)
wspr['type'] = wspr['time.1'].str[4]
```

```python
wspr
```

```python
wspr['alt'] = wspr['power']
```

```python
wspr.shape
```

```python
wspr_0 = wspr[wspr.type == '0']
wspr_0.shape
```

```python
wspr_8 = wspr[wspr.type == '8']
wspr_8.shape
```

```python
#df.iloc[:,0]  *= -1
wspr_0.loc[:,'alt'] *= 20 
```

```python
wspr_0
```

```python
wspr_8.loc[:,'alt'] *= 300 
```

```python
wspr_8
```

```python
frames = [wspr_0, wspr_8]
new_wspr = pd.concat(frames)
new_wspr.shape
```

```python
new_wspr.sort_index(inplace=True)
new_wspr
```

```python
new_wspr.to_csv('new_wspr.csv', index = False)
```

```python
# I only print for type 0 packets and add the 0 detail to type 8 base. 
with open("new_wspr.csv", "r") as a_file:
  for line in a_file:
    spline = line.split(',')
    #print(spline[26])
    if spline[25] == '8':
      base_altitude = int(spline[26])
    elif spline[25] == '0':
      altitude = base_altitude + int(spline[26])
      print(spline[1], altitude)
      
        

    
```
